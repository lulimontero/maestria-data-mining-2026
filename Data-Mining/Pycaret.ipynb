```python
%%writefile pycaret_full_test.py

from pycaret.datasets import get_data
from pycaret.regression import RegressionExperiment

print("1. Import OK")

df = get_data("insurance")

print("2. Dataset OK")
print(df.head())

exp_reg = RegressionExperiment()

print("3. Experiment OK")

reg = exp_reg.setup(
    data=df,
    target="charges"
)

print("4. Setup OK")

best_reg = reg.compare_models()

print("5. Compare OK")
print(best_reg)
```

```python
%%writefile pycaret_analysis.py

from pycaret.datasets import get_data
from pycaret.regression import RegressionExperiment
from pycaret.classification import ClassificationExperiment


# ======================
# REGRESSION
# ======================

print("\n===== REGRESSION =====\n")

df_reg = get_data("insurance")

exp_reg = RegressionExperiment()

reg = exp_reg.setup(
    data=df_reg,
    target="charges"
)

best_reg = reg.compare_models()

print("\nBest regression model:")
print(best_reg)


# ======================
# CLASSIFICATION
# ======================

print("\n===== CLASSIFICATION =====\n")

df_clf = get_data("hepatitis")

exp_clf = ClassificationExperiment()

clf = exp_clf.setup(
    data=df_clf,
    target="Class",
    imputation_type="iterative",
    categorical_iterative_imputer="rf",
    numeric_iterative_imputer="rf",
    iterative_imputation_iters=5
)

print("\nTraining data:")
print(clf.X_train.head())

best_clf = clf.compare_models(sort="AUC")

print("\nBest classification model:")
print(best_clf)
```
