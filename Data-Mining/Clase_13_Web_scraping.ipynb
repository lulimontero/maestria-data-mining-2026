# Web scraping

Leemos páginas web para sacar datos estructurados

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36"
}

res = requests.get("https://quelibroleo.com/los-tres-mundos", headers=headers)

if res.status_code != 200:
    print(res.status_code)

print(res.content)
```

## Beautiful Soup

```python
tree = BeautifulSoup(res.content, 'html.parser')
print(tree.prettify()[:400])
```

```python
tree.find(class_="card-block").find("a").get_text()
```

```python
tree.find_all("h3")[1].get_text()
```
