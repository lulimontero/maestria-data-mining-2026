# Lectura y escritura de datos

```python
df = pd.read_csv("https://data.insideairbnb.com/argentina/ciudad-aut%C3%B3noma-de-buenos-aires/buenos-aires/2026-01-25/visualisations/listings.csv", index_col="id", parse_dates=["last_review"])
df.head()
```

```python
df.dtypes
```

# Limpieza de columnas

```python
df["name"] = df["name"].str.replace("★", "").str.replace("Depto", "Departamento")
```

```python
df["license"] = df["license"].fillna("no license")
#df["license"].fillna("no license", inplace=True)
```

```python
# quitar columnas: 2 formas equivalentes

df = df.drop(columns=["license"])
#df = df[df.columns.drop("license")]
```

# Discretización

```python
df["minimum_nights"].describe()
```

```python
# bins de igual longitud (q)
pd.qcut(df["minimum_nights"], q=3, labels=['short', 'medium', 'long'])#.value_counts()
```

```python
# bins a mano
pd.cut(df["minimum_nights"], bins=[df["minimum_nights"].min(), 2, 4, df["minimum_nights"].max()], labels=['short', 'medium', 'long']).value_counts()
```

# Numeración
