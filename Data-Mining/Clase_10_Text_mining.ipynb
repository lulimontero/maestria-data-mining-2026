# Text Mining

```python
import pandas as pd
```

## Lectura de datos

```python
df = pd.read_excel('comentarios_parrilas.csv.xlsx')
df.shape
```

```python
df.head()
```

```python
df = df.sample(100)
```

## Minúsculas

```python
df['comentario'] = df['comentario'].str.lower()
df.head()
```

## Spacy

```python
import spacy

import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
! spacy download es_core_news_sm
```

```python
nlp = spacy.load("es_core_news_sm")
```

## Tokenizador

```python
doc = nlp(df.iloc[3].comentario)
print(doc.text)
for token in doc:
    print(token.text, "-->", token.pos_)
```

```python
df['comentario_tok'] = df['comentario'].apply(lambda texto: [token.text for token in nlp(texto)])
df['comentario_tok']
```

## Lematizador

```python
doc = nlp(df.iloc[2].comentario)
print(doc.text)
for token in doc:
    print(token.text, "-->", token.lemma_)
```

```python
df['comentario_tok'] = df['comentario'].apply(lambda texto: [token.lemma_ for token in nlp(texto)])
df['comentario_tok']
```

## Normalización

```python
m = {'morci': 'morcilla', 'chori': 'chorizo', 'muuuuy': 'muy'}

df['comentario_tok'] = df['comentario_tok'].apply(lambda texto: [m.get(token, token) for token in texto])
df['comentario_tok']
```

```python
df['comentario_tok'] = df['comentario_tok'].apply(lambda texto: [token for token in texto if token not in (":", ",", ".", "!")])
df['comentario_tok']
```

## Stopwords

```python
stopwords = nlp.Defaults.stop_words
stopwords
```

```python
stopwords.update(['es', 'osea', 'sólo', 'la', 'el', 'los', 'las'])

df['comentario_tok'] = df['comentario_tok'].apply(lambda texto: [token for token in texto if token not in stopwords])
df['comentario_tok']
```

## Preparación para predicción

```python
df['comentario'] = df['comentario_tok'].str.join(' ')
df['comentario']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1, 3), min_df=0.005, max_df=0.7, strip_accents='ascii', max_features=2000)
vectorized_data = vectorizer.fit_transform(df['comentario'])
vectorized_data.toarray()
```

```python
vectorizer.get_feature_names_out().tolist()
```

```python
count_df = pd.DataFrame(vectorized_data.toarray(), columns=vectorizer.get_feature_names_out(), index=df.index)
count_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0.001, max_df=0.9, strip_accents='ascii', max_features=2000)
vectorized_data = vectorizer.fit_transform(df['comentario'])
```

```python
vectorizer.get_feature_names_out()
```

```python
tfidf_df = pd.DataFrame(vectorized_data.toarray(), columns=vectorizer.get_feature_names_out(), index=df.index)
tfidf_df
```
